```python
import pandas as pd
import re as r
```

```python
df = pd.read_csv("final.csv")
```

```python
df.head()
```

```python
comments = df.iloc[:, 0].values
senti = df.iloc[:, 1].values
```

```python
comments
```

```python
processed_comments=[]

for text in range(0,len(comments)):
    
    processed_comment = r.sub(r'\W', ' ', str(comments[text]))
    
    processed_comment= r.sub(r'\s+[a-zA-Z]\s+', ' ', processed_comment)

    # Remove single characters from the start
    processed_comment= r.sub(r'\^[a-zA-Z]\s+', ' ', processed_comment) 

    # Substituting multiple spaces with single space
    processed_comment = r.sub(r'\s+', ' ', processed_comment, flags=r.I)

    # Removing prefixed 'b'
    processed_comment = r.sub(r'^b\s+', '', processed_comment)

    # Converting to Lowercase
    processed_comment = processed_comment.lower()

    processed_comments.append(processed_comment)
```

```python
processed_comments[:10]
```

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer (max_features=5000, min_df=7, max_df=0.8, stop_words=stopwords.words('english'))

vectorizer_save = vectorizer.fit(processed_comments)

import joblib
joblib.dump(vectorizer_save,'vectorized.pkl')
processed_comments = vectorizer.transform(processed_comments).toarray()
```

```python
import sklearn
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,x_test ,Y_train ,y_test = train_test_split(processed_comments,senti,test_size =0.2 ,random_state=45)#25
```

```python
X_train.shape
```

```python
#from xgboost import XGBClassifier
#from sklearn.metrics import accuracy_score
```

```python
model = RandomForestClassifier(n_estimators = 1000,bootstrap=True) # bootstrap=true,n=1000
```

```python
model.fit(X_train,Y_train)
```

```python
predict_senti = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,predict_senti))
print(classification_report(y_test,predict_senti))
print(accuracy_score(y_test, predict_senti))
```

```python
 from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
clf1 = AdaBoostClassifier(n_estimators=1000) # 0.77
clf1.fit(X_train, Y_train)
y_pred = clf1.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
clf = GradientBoostingClassifier(n_estimators=800) #0.851
clf.fit(X_train, Y_train)
y_pred = clf.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```
